```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime
import warnings

warnings.filterwarnings('ignore')
```

```python
# European Tickers (Italy, France, Germany, Spain, the Netherlands), last 5 years
european_tickers = ['ENEL.MI', 'ISP.MI', 'ENI.MI','AIR.PA', 'MC.PA', 'OR.PA','SAP.DE', 'SIE.DE', 'ALV.DE','IBE.MC', 'SAN.MC','ASML.AS']

end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=5*365)

print(f"Downloading data from {start_date} to {end_date}")
```

```python
# Downloading data
price_data = yf.download(european_tickers, start=start_date, end=end_date)

# Checking for NaN
print("NaN values for ticket")
print(price_data.isna().sum())

# Filling NaNs
price_data = price_data.fillna(method='ffill')

price_data.head()
```

```python
import pandas as pd

# ESG Risk scores (Sustainalytics via Morningstar)
# Lower score = Better (lower ESG risk)

esg_scores_dict = {
    'ENEL.MI': 17.39,  # Utilities
    'ISP.MI': 8.09,   # Finance
    'ENI.MI': 24.91,   # Energy (Oil & Gas)
    'AIR.PA': 20.33,   # Aerospace
    'MC.PA': 13.55,    # Luxury (LVMH)
    'OR.PA': 16.90,    # Consumer goods (L'Oreal)
    'SAP.DE': 11.23,    # Tech 
    'SIE.DE': 22.94,   # Automation and digitalization 
    'ALV.DE': 13.16,   # Assicurazioni
    'IBE.MC': 20.32,   # Utilities 
    'SAN.MC': 9.89,   # Finance
    'ASML.AS': 8.93   # Tech
}

print("Creating hardcoded ESG dataset...")

# Converting in Pandas DataFrame
esg_df = pd.DataFrame(list(esg_scores_dict.items()), columns=['Ticker', 'ESG_Risk_Score'])
esg_df.set_index('Ticker', inplace=True)

display(esg_df)
```

```python
# Saving in .CSV format
price_data.to_csv('european_prices_5Y.csv')
esg_df.to_csv('european_esg_scores.csv')

print("Data saved!")
```
